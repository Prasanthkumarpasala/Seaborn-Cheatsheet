# SEABORN FULL CHEATSHEET - FULL OPERATIONS

```python
# Ensures plots are rendered inline in Jupyter Notebook

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

print("Matplotlib + Seaborn ready. Plots will display below.")
```

# LOAD DATASETS

```python
# Load built-in dataset
df = sns.load_dataset("tips")

print("Dataset loaded: tips")
print(df.head())
```

# SEABORN THEMES & STYLES

```python
# Set overall visual theme
sns.set_theme()

print("Default seaborn theme applied")

# Change plot style
sns.set_style("whitegrid")

print("Style set to whitegrid")

# Available styles:

# darkgrid

# whitegrid

# dark

# white

# ticks
```

# SCATTER PLOT (VISIBLE OUTPUT)

```python
# Scatter plot = relationship between two numeric variables
sns.scatterplot(data=df, x="total_bill", y="tip")

plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()

```

# LINE PLOT

```python
# Line plot = trend over continuous variable
sns.lineplot(data=df, x="size", y="tip")

plt.title("Line Plot: Tip vs Size")
plt.show()

```

# RELPLOT (FIGURE-LEVEL)

```python
# relplot = high-level interface (creates its own figure)
sns.relplot(data=df, x="total_bill", y="tip", kind="scatter")

plt.show()


# relplot = high-level interface (creates its own figure)
sns.relplot(data=df, x="total_bill", y="tip", kind="line")

plt.show()
```

# BAR PLOT

```python
# Bar plot = mean value per category
sns.barplot(data=df, x="day", y="total_bill")

plt.title("Average Total Bill per Day")
plt.show()
```

# COUNT PLOT

```python
# Count plot = frequency of categories
sns.countplot(data=df, x="day")

plt.title("Count of Records per Day")
plt.show()
```

# BOX PLOT

```python
# Box plot = distribution + outliers
sns.boxplot(data=df, x="day", y="total_bill")

plt.title("Box Plot of Total Bill by Day")
plt.show()
```

# VIOLIN PLOT

```python
# Violin plot = distribution shape + boxplot
sns.violinplot(data=df, x="day", y="total_bill")

plt.title("Violin Plot")
plt.show()
```

# STRIP PLOT

```python
# Strip plot = individual data points
sns.stripplot(data=df, x="day", y="total_bill")

plt.title("Strip Plot")
plt.show()
```

# SWARM PLOT

```python
# Swarm plot = non-overlapping strip plot
sns.swarmplot(data=df, x="day", y="total_bill")

plt.title("Swarm Plot")
plt.show()
```

# CATPLOT

```python
# catplot = figure-level categorical plot
sns.catplot(data=df, x="day", y="total_bill", kind="bar")

plt.show()

# catplot = figure-level categorical plot
sns.catplot(data=df, x="day", y="total_bill", kind="box")

plt.show()
```

# HISTOGRAM

```python
# Histogram = frequency distribution
sns.histplot(data=df, x="total_bill", bins=20)

plt.title("Histogram of Total Bill")
plt.show()
```

# KDE PLOT

```python
# KDE plot = probability density curve
sns.kdeplot(data=df, x="total_bill", fill=True)

plt.title("KDE Plot")
plt.show()
```

# RUG PLOT

```python
# Rug plot = small ticks for each observation
sns.rugplot(data=df, x="total_bill")

plt.title("Rug Plot")
plt.show()
```

# JOINT PLOT

```python
# Joint plot = scatter + histograms
sns.jointplot(data=df, x="total_bill", y="tip")

plt.show()
```

# PAIR PLOT

```python
# Pair plot = relationships between all numeric columns
sns.pairplot(df)

plt.show()
```

# REGRESSION PLOT

```python
# Regression plot = scatter + best-fit line
sns.regplot(data=df, x="total_bill", y="tip")

plt.title("Regression Plot")
plt.show()
```

# LMPLOT

```python
# lmplot = regression with grouping (figure-level)
sns.lmplot(data=df, x="total_bill", y="tip", hue="sex")

plt.show()

```

# HEATMAP

```python
# Correlation heatmap
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

# CLUSTERMAP

```python
# Clustered heatmap
sns.clustermap(corr)

plt.show()
```

# FACETGRID

```python
# FacetGrid = multiple plots by category
g = sns.FacetGrid(df, col="sex")
g.map(sns.histplot, "total_bill")

plt.show()
```

# HUE / SIZE / STYLE

```python
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="sex",
    size="size"
)

plt.title("Hue + Size Encoding")
plt.show()
```
